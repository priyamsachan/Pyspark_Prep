#### **Mastering collect_list and Aggregation in PySpark: Aggregating Item Weights by Name!**

```python
data = [
    ("john", "tomato", 2),
    ("ğš‹ğš’ğš•ğš•", "ğšŠğš™ğš™ğš•ğš", 2),
    ("john", "ğš‹ğšŠğš—ğšŠğš—ğšŠ", 2),
    ("john", "tomato", 3),
    ("ğš‹ğš’ğš•ğš•", "ğšğšŠğšŒğš˜", 2),
    ("ğš‹ğš’ğš•ğš•", "ğšŠğš™ğš™ğš•ğš", 2),
]
schema = "name string,item string,weight int"
df = spark.createDataFrame(data, schema)
display(df)
```

```python
from pyspark.sql import functions as F
df2 = df.groupBy("name","item").agg(F.sum("weight").alias("sum_weight"))
df_final=df2.groupBy("name").agg(F.collect_list(F.struct("item","sum_weight")).alias("Items"))
display(df_final)
```
