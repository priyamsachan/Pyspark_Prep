### **Pyspark code to get list & display all dataframes in the Notebook?**

```python
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, FloatType

# Initialize Spark session
spark = SparkSession.builder.appName("SampleSalesData").getOrCreate()

# Define the schema for the DataFrame
schema = StructType([
    StructField("SaleID", IntegerType(), True),
    StructField("Product", StringType(), True),
    StructField("Quantity", IntegerType(), True),
    StructField("Price", FloatType(), True)
])

# Sample data
data = [
    (1, "ProductA", 10, 20.5),
    (2, "ProductB", 5, 15.0),
    (3, "ProductC", 7, 30.0),
    (4, "ProductD", 3, 12.5),
    (5, "ProductE", 8, 50.0),
    (6, "ProductF", 2, 60.0),
    (7, "ProductG", 1, 100.0),
    (8, "ProductH", 4, 25.0),
    (9, "ProductI", 6, 10.0),
    (10, "ProductJ", 9, 5.0)
]

# Create DataFrame
df = spark.createDataFrame(data, schema)

# Show the DataFrame
df.show()
```

```python
# Initialize Spark session (if not already initialized)
spark = SparkSession.builder.appName("SampleProductsData").getOrCreate()

# Define the schema for the DataFrame
product_schema = StructType([
    StructField("ProductID", IntegerType(), True),
    StructField("ProductName", StringType(), True),
    StructField("Price", FloatType(), True)
])

# Sample data
product_data = [
    (1, "ProductA", 20.5),
    (2, "ProductB", 15.0),
    (3, "ProductC", 30.0),
    (4, "ProductD", 12.5),
    (5, "ProductE", 50.0)
]

# Create DataFrame
product_df = spark.createDataFrame(product_data, product_schema)

# Show the DataFrame
product_df.show()
```
